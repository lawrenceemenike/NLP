```python
# Majors with the Highest Potential
highest_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
```

```python
highest_potential
```

```python
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# Calculate the spread
clean_df['Spread'] = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']

# Majors with Greatest Spread in Salaries
highest_spread = clean_df.sort_values('Spread', ascending=False)
highest_spread[['Undergraduate Major', 'Spread']].head()
```

```python
# Group categories
clean_df.groupby('Group').count()
```

```python
# Check data types of each column
print(clean_df.dtypes)

# If 'Group' column is of type object (string), it might be the issue.
# If so, you might need to reconsider your grouping strategy or convert
# the 'Group' column to a numerical representation if it makes sense in your context.

# If other columns you want to aggregate have string values, you need to:
# 1. Identify those columns.
# 2. Decide if those string values are erroneous and need cleaning,
#    or if those columns should be excluded from the mean calculation.

# Example of excluding non-numeric columns from the mean calculation:
numeric_df = clean_df.select_dtypes(include=['number'])
group_means = numeric_df.groupby('Group').mean()
print(group_means)
```
